```python
numerical_cols = train_data.select_dtypes(exclude='O').columns
categorical_cols = train_data.select_dtypes(include='O').columns.drop('addons')
```

```python
print(numeric_cols)
print(categorical_cols)
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(handle_unknown='ignore', sparse=False)

# Fit and transform the OneHotEncoder using the categorical columns of train_data
one_hot_encoded_cols = pd.DataFrame(ohe.fit_transform(train_data[categorical_cols]))

# OneHotEncoder removes column names; put them back
one_hot_encoded_cols.columns = ohe.get_feature_names_out(categorical_cols)

# OneHotEncoder changes the index; put it back
one_hot_encoded_cols.index = train_data.index

# Drop categorical columns from the original DataFrame
train_data = train_data.drop(categorical_cols, axis=1)

train_data = pd.concat([train_data, one_hot_encoded_cols], axis=1)
```

```python
train_data = pd.concat([train_data, one_hot_encoded_cols], axis=1)
train_data.drop('addons', axis=1, inplace=True)
```

```python
# Split the data into features (X) and target variable (y)  
X = train_data.drop(['id', 'pret'], axis=1)
y = train_data['pret']

# Split the data into training and validation sets
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

# Define the model
model = RandomForestRegressor(n_estimators=100, random_state=0)

# Fit the model (will take about a minute to train)
model.fit(train_X, train_y)

# Get the mean absolute error on the validation data
preds = model.predict(val_X)
print(mean_squared_error(val_y, preds, squared=False))
```
